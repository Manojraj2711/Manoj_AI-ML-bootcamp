```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy as s

df = pd.read_csv('/content/drive/MyDrive/wig nsoh new.txt')
```

```python
first=df.iloc[:8,]
second=df.iloc[9:17,]
third=df.iloc[18:26,]
fourth=df.iloc[27:35,]
fifth=df.iloc[36:44,]
sixth=df.iloc[45:53,]
seventh=df.iloc[54:62,]
eigth=df.iloc[63:71,]
ninth=df.iloc[72:80,]
tenth=df.iloc[81:89,]
eleventh=df.iloc[90:98,]
```

```python
first=first.astype(float)
second=second.astype(float)
third=third.astype(float)
fourth=fourth.astype(float)
fifth=fifth.astype(float)
sixth=sixth.astype(float)
seventh=seventh.astype(float)
eigth=eigth.astype(float)
ninth=ninth.astype(float)
tenth=tenth.astype(float)
eleventh=eleventh.astype(float)
```

```python
L=[]
L.append(first)
L.append(second)
L.append(third)
L.append(fourth)
L.append(fifth)
L.append(sixth)
L.append(seventh)
L.append(eigth)
L.append(ninth)
L.append(tenth)
L.append(eleventh)
```

```python
ax=plt.axes(projection='3d')
area = np.zeros(11)
c = 0
for i in L:
  ax.plot3D(i.x,i.y,i.z)
  ax.plot3D(i.x,-1*i.y,i.z)
  area[c] = 2*s.integrate.simpson(i.y, i.z)
  print(area[c])
  c = c+1


plt.show()
```

```python
vol = s.integrate.simpson(area,dx=10)
print(vol)
```
